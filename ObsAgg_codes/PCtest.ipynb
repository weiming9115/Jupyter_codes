```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

```python
X.shape
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
print(pca.explained_variance_)
```

```python
pc1 = np.matrix(pca.components_[:,0])
pc2 = np.matrix(pca.components_[:,1])
```

```python
A = np.matrix(np.dot(X,pc1.T))
B = np.matrix(np.dot(X,pc2.T))
```

```python
mode1 = np.matrix.dot(A,pc1)
mode2 = np.matrix.dot(B,pc2)
```

```python
plt.scatter(X[:, 0], X[:, 1],color='k')
plt.scatter(mode1.A[:,0]+mode2.A[:,0],mode1.A[:,1]+mode2.A[:,1],color='b')
#plt.scatter(mode2.A[:,0],mode2.A[:,1],color='r')
```

```python
plt.plot(X[:,0])
plt.plot(mode1.A[:,0]+mode2.A[:,0])
```

```python
plt.plot(X[:,1])
plt.plot(mode1.A[:,1]+mode2.A[:,1])
```
